### Model Assumptions

The reliability of the linear regression model depends on the following assumptions:

1. **Linearity**: The relationship between GDP and energy consumption is linear.
2. **Independence**: Observations are independent, meaning there is no systematic dependence among countries.
3. **Homoscedasticity**: The variance of residuals remains constant across all levels of GDP.
4. **Normality of Errors**: Residuals are normally distributed.
5. **No Multicollinearity**: There is no perfect correlation among independent variables. Since the model includes only GDP, this assumption is automatically satisfied.
6. **Zero Conditional Mean Assumption (E[u|X] = 0)**: The expected value of the error term, conditional on GDP (X), is zero. This assumption ensures that the OLS estimates of the coefficients ($\beta_0$ and $\beta_1$) are unbiased. In other words, the factors not captured by GDP (such as energy policies or technological advancements) should not systematically influence energy consumption in a way that correlates with GDP.

#### 2.1 Linearity

We assess the linearity assumption by plotting the observed energy consumption against GDP and overlaying the fitted regression line.


```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import statsmodels.stats.api as sms
import json
from typing import Any, Dict


def load_data(file_path: str) -> pd.DataFrame:
    """
    Load data from a JSON file into a DataFrame.

    Args:
        file_path (str): The path to the JSON file containing the data.

    Returns:
        pd.DataFrame: The data as a pandas DataFrame.
    """
    with open(file_path, 'r') as file:
        data_json = json.load(file)
    return pd.DataFrame(data_json)


def fit_regression_model(df: pd.DataFrame, independent_variable: str, dependent_variable: str) -> sm.OLS:
    """
    Fit a linear regression model and return the model object.

    Args:
        df (pd.DataFrame): The dataframe containing the data.
        independent_variable (str): The name of the independent variable.
        dependent_variable (str): The name of the dependent variable.

    Returns:
        sm.OLS: The fitted regression model object.
    """
    X = sm.add_constant(df[independent_variable])  # Adding constant for intercept
    y = df[dependent_variable]
    model = sm.OLS(y, X).fit()
    return model


def plot_regression_line(df: pd.DataFrame, model: sm.OLS, independent_variable: str, dependent_variable: str) -> None:
    """
    Plot scatter plot and regression line.

    Args:
        df (pd.DataFrame): The dataframe containing the data.
        model (sm.OLS): The fitted regression model.
        independent_variable (str): The name of the independent variable.
        dependent_variable (str): The name of the dependent variable.

    Returns:
        None
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=independent_variable, y=dependent_variable, data=df, color='blue', label='Data points')

    # Plot the regression line
    plt.plot(df[independent_variable], model.fittedvalues, color='red', label='Regression Line')

    # Labels and title
    plt.title(f'{independent_variable} vs {dependent_variable}')
    plt.xlabel(f'{independent_variable} (euros)')
    plt.ylabel(f'{dependent_variable} (kWh)')
    plt.legend()
    plt.show()


def plot_correlation_matrix(df: pd.DataFrame) -> None:
    """
    Plot the correlation matrix using seaborn's heatmap.

    Args:
        df (pd.DataFrame): The dataframe containing the data.

    Returns:
        None
    """
    correlation_matrix = df.corr()

    plt.figure(figsize=(6, 4))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
    plt.title('Correlation Matrix')
    plt.show()

    # Print the correlation matrix for review
    print("\nCorrelation Matrix:")
    print(correlation_matrix)


# Main execution flow
if __name__ == "__main__":
    # Load the dataset
    euroenergy = load_data('data.json')

    # Fit the regression model
    model = fit_regression_model(euroenergy, 'gdp', 'energy')

    # Plot the regression line
    plot_regression_line(euroenergy, model, 'gdp', 'energy')

    # Plot the correlation matrix
    plot_correlation_matrix(euroenergy)


```

The scatterplot and regression line provide a visual check for linearity. A strong alignment between the data points and the regression line confirms the linearity assumption.

#### 2.2 Homoscedasticity

To check for constant variance in residuals, we conduct the Breusch-Pagan test.

```python
import statsmodels.api as sm
import statsmodels.stats.api as sms
import pandas as pd

# Function to fit the OLS model
def fit_ols_model(independent_variable: pd.Series, dependent_variable: pd.Series):
    """
    Fit a simple OLS regression model.

    Parameters:
    independent_variable (pd.Series): The independent variable (e.g., GDP).
    dependent_variable (pd.Series): The dependent variable (e.g., energy consumption).

    Returns:
    statsmodels.regression.linear_model.OLSResults: The fitted OLS model.
    """
    X = sm.add_constant(independent_variable)  # Adding constant (intercept) to the independent variable
    model = sm.OLS(dependent_variable, X).fit()
    return model

# Function to perform the Breusch-Pagan test for heteroscedasticity
def breusch_pagan_test(model):
    """
    Perform the Breusch-Pagan test for heteroscedasticity.

    Parameters:
    model (statsmodels.regression.linear_model.OLSResults): The fitted OLS model.

    Returns:
    tuple: A tuple containing the LM statistic, p-value, F-statistic, and F-p-value.
    """
    bp_test = sms.het_breuschpagan(model.resid, model.model.exog)
    return bp_test

# Function to load the dataset from a JSON file
def load_dataset(file_path: str):
    """
    Load the dataset from a JSON file.

    Parameters:
    file_path (str): The path to the JSON file.

    Returns:
    pd.DataFrame: The loaded dataset.
    """
    return pd.read_json(file_path)

# Example usage and integration test
def run_breusch_pagan_test_on_data(file_path: str):
    """
    Run the OLS model and Breusch-Pagan test on a given dataset.

    Parameters:
    file_path (str): The path to the JSON file containing the 'gdp' and 'energy' columns.

    Returns:
    None: This function prints the results of the test.
    """
    # Load the dataset
    euroenergy = load_dataset(file_path)
    
    # Check if the required columns are present in the dataset
    if 'gdp' not in euroenergy.columns or 'energy' not in euroenergy.columns:
        print("Error: Dataset must contain 'gdp' and 'energy' columns.")
        return
    
    # Fit the OLS model
    model = fit_ols_model(euroenergy['gdp'], euroenergy['energy'])
    
    # Perform the Breusch-Pagan test
    bp_test_results = breusch_pagan_test(model)
    
    # Display the results
    print("Breusch-Pagan test results:")
    print("LM statistic:", bp_test_results[0])
    print("p-value:", bp_test_results[1])
    print("F-statistic:", bp_test_results[2])
    print("f-p-value:", bp_test_results[3])

# Run the test on the JSON file
run_breusch_pagan_test_on_data("data.json")
```

When the test $p$-value is greater than 0.05, we fail to reject the null hypothesis of homoscedasticity, confirming constant variance.

## Reflection on Limitations of the Analysis

In this analysis, we have examined the relationship between GDP per capita and energy consumption using a simple linear regression model. While the results suggest a statistically significant positive relationship, there are several limitations that need to be addressed to ensure the robustness and validity of the analysis.

### 1. Validity of the Zero Conditional Mean Assumption

The key assumption of linear regression models is the zero conditional mean assumption, which states that the error terms (residuals) have an expected value of zero, conditional on the explanatory variable. In other words, the error terms must be uncorrelated with GDP per capita. If this assumption is violated, the model may yield biased estimates of the regression coefficients.

##### Potential Violations:

-   **Omitted Variable Bias:** If there are other important variables affecting energy consumption (such as population size, energy efficiency, or energy mix) that are not included in the model, the residuals may capture the effect of these missing factors, violating the zero conditional mean assumption. This could lead to biased estimates of the relationship between GDP and energy consumption.

-   **Measurement Errors:** If there are errors in the measurement of GDP or energy consumption, this could cause errors in the residuals, affecting the assumptions underlying the regression model.

-   **Endogeneity:** If energy consumption somehow affects GDP (for example, energy demand driving economic growth), this could lead to endogeneity and bias the estimates. In such cases, the direction of causality is not clear, and the model would not properly estimate the effect of GDP on energy consumption.

### 2. Model Specification

The model assumes a linear relationship between GDP per capita and energy consumption. While this might hold in the dataset, it is possible that the relationship is non-linear or that there are diminishing returns to energy consumption as GDP increases. A more complex model, such as a log-log model or a quadratic model, might capture these dynamics better.

Additionally, the Intercept term in the model is not statistically significant, which may indicate that the model does not fully explain the behavior of energy consumption at very low levels of GDP. This suggests that there may be other factors influencing energy consumption at lower income levels that are not accounted for in the current model.

```python
import statsmodels.api as sm
import pandas as pd
import json
from typing import Dict
from statsmodels.stats.outliers_influence import variance_inflation_factor

def load_json_data(file_path: str) -> Dict:
    """
    Loads JSON data from a file and returns it as a dictionary.

    Parameters:
    file_path (str): The path to the JSON file.

    Returns:
    Dict: A dictionary containing the data loaded from the JSON file.
    """
    with open(file_path, 'r') as file:
        data_json = json.load(file)
    return data_json

def convert_to_dataframe(data: Dict) -> pd.DataFrame:
    """
    Converts a dictionary into a pandas DataFrame.

    Parameters:
    data (Dict): A dictionary containing the data.

    Returns:
    pd.DataFrame: A DataFrame containing the data.
    """
    return pd.DataFrame(data)

def prepare_independent_variables(euroenergy: pd.DataFrame) -> pd.DataFrame:
    """
    Prepares the independent variables by extracting relevant columns and adding a constant term.

    Parameters:
    euroenergy (pd.DataFrame): The DataFrame containing the dataset.

    Returns:
    pd.DataFrame: The independent variables DataFrame with a constant term.
    """
    X = euroenergy[['gdp', 'population']]
    X = sm.add_constant(X)  # Adds a constant term to the independent variables
    return X

def prepare_dependent_variable(euroenergy: pd.DataFrame) -> pd.Series:
    """
    Prepares the dependent variable by extracting the relevant column.

    Parameters:
    euroenergy (pd.DataFrame): The DataFrame containing the dataset.

    Returns:
    pd.Series: The dependent variable (Energy consumption).
    """
    return euroenergy['energy']

def fit_ols_model(X: pd.DataFrame, y: pd.Series):
    """
    Fits an Ordinary Least Squares (OLS) regression model.

    Parameters:
    X (pd.DataFrame): The independent variables with a constant term.
    y (pd.Series): The dependent variable.

    Returns:
    statsmodels.regression.linear_model.OLSResults: The fitted OLS regression model results.
    """
    model = sm.OLS(y, X).fit()
    return model

def display_regression_results(model) -> None:
    """
    Displays the summary of the regression model.

    Parameters:
    model (statsmodels.regression.linear_model.OLSResults): The fitted OLS regression model.

    Returns:
    None: This function prints the summary of the regression results.
    """
    print(model.summary())

def calculate_vif(X: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate Variance Inflation Factor (VIF) for each independent variable in the model.
    
    Parameters:
    X (pd.DataFrame): The independent variables DataFrame.
    
    Returns:
    pd.DataFrame: A DataFrame containing the VIF values for each variable.
    """
    vif_data = pd.DataFrame()
    vif_data['Variable'] = X.columns
    
    # Calculate VIF for each variable using statsmodels
    vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return vif_data

def run_regression_analysis(file_path: str) -> None:
    """
    Loads data, prepares variables, fits an OLS regression model, calculates VIF, and displays the results.

    Parameters:
    file_path (str): The path to the JSON file containing the dataset.

    Returns:
    None: This function runs the regression analysis and displays the results.
    """
    # Load and prepare the dataset
    data_json = load_json_data(file_path)
    euroenergy = convert_to_dataframe(data_json)
    
    # Prepare independent and dependent variables
    X = prepare_independent_variables(euroenergy)
    y = prepare_dependent_variable(euroenergy)
    
    # Fit the OLS model
    model = fit_ols_model(X, y)
    
    # Display the regression results
    display_regression_results(model)
    
    # Calculate and display VIF
    vif_data = calculate_vif(X)
    print("\nVariance Inflation Factor (VIF) for each variable:")
    print(vif_data)

# Example usage: run the regression analysis and VIF calculation on the given dataset
run_regression_analysis('json2.json')
```

```python
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_residuals_vs_fitted(model: sm.regression.linear_model.RegressionResults, ax: plt.Axes) -> None:
    """
    Plots the residuals versus fitted values to check for homoscedasticity.
    
    Parameters:
    model (sm.regression.linear_model.RegressionResults): The fitted OLS model from which residuals are taken.
    ax (plt.Axes): The axes to plot the graph on.

    Returns:
    None: This function directly plots on the provided axes.
    """
    residuals = model.resid
    sns.scatterplot(x=model.fittedvalues, y=residuals, ax=ax)
    ax.set_title('Residuals vs Fitted')
    ax.set_xlabel('Fitted Values')
    ax.set_ylabel('Residuals')

def plot_qq(residuals: np.ndarray, ax: plt.Axes) -> None:
    """
    Plots a Q-Q plot of the residuals to check for normality.

    Parameters:
    residuals (np.ndarray): The residuals from the fitted OLS model.
    ax (plt.Axes): The axes to plot the Q-Q plot on.

    Returns:
    None: This function directly plots on the provided axes.
    """
    sm.qqplot(residuals, line='45', ax=ax)
    ax.set_title('Q-Q Plot')

def visualize_residuals(model: sm.regression.linear_model.RegressionResults) -> None:
    """
    Visualizes the residuals of the fitted OLS model by plotting:
    - Residuals vs Fitted values to check for homoscedasticity.
    - Q-Q plot to check for normality of residuals.

    Parameters:
    model (sm.regression.linear_model.RegressionResults): The fitted OLS model.

    Returns:
    None: This function creates and displays the plots.
    """
    residuals = model.resid

    # Create subplots
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Plot residuals vs fitted values
    plot_residuals_vs_fitted(model, axs[0])

    # Plot Q-Q plot
    plot_qq(residuals, axs[1])

    plt.tight_layout()
    plt.show()

# Convert to DataFrame
euroenergy = pd.DataFrame(data)

# Prepare the independent variables (X) and dependent variable (y)
X = euroenergy[['gdp', 'population']]
X = sm.add_constant(X)  # Adds constant term
y = euroenergy['energy']

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Visualize the residuals
visualize_residuals(model)
```

```python
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

def log_transform_energy(data):
    """
    Apply a log transformation to the 'energy' column in the dataset.
    
    Parameters:
    data (pandas.DataFrame): The dataset containing the 'energy' column.
    
    Returns:
    pandas.DataFrame: The dataset with a new 'log_energy' column.
    """
    data['log_energy'] = np.log(data['energy'])
    return data

def fit_log_regression(data):
    """
    Fit a log-transformed OLS regression model using GDP and population as independent variables.
    
    Parameters:
    data (pandas.DataFrame): The dataset containing the necessary columns.
    
    Returns:
    sm.OLS: The fitted OLS model.
    pandas.Series: The residuals of the fitted model.
    """
    X = data[['gdp', 'population']]  # Independent variables
    X = sm.add_constant(X)  # Add constant term for the intercept
    y = data['log_energy']  # Transformed dependent variable

    log_model = sm.OLS(y, X).fit()
    log_residuals = log_model.resid
    return log_model, log_residuals

def plot_residuals(log_model, log_residuals):
    """
    Plot residuals vs fitted values and a Q-Q plot for residuals.
    
    Parameters:
    log_model (sm.OLS): The fitted OLS regression model.
    log_residuals (pandas.Series): The residuals from the fitted model.
    """
    # Plot residuals vs fitted values
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    sns.scatterplot(x=log_model.fittedvalues, y=log_residuals, ax=axs[0])
    axs[0].set_title('Residuals vs Fitted (Log-Transformed)')
    axs[0].set_xlabel('Fitted Values')  # Set x-axis label
    axs[0].set_ylabel('Residuals')      # Set y-axis label

    # Q-Q plot (for Normality of Residuals)
    sm.qqplot(log_residuals, line='45', ax=axs[1])
    axs[1].set_title('Q-Q Plot (Log-Transformed)')

    plt.tight_layout()
    plt.show()

# Example usage
data = log_transform_energy(data)  # Step 1: Log-transform the 'energy' column
log_model, log_residuals = fit_log_regression(data)  # Step 2: Fit the regression model
plot_residuals(log_model, log_residuals)  # Step 3: Plot residuals and Q-Q plot
```

**Residuals vs Fitted Plot (Log-Transformed)**: The scattered points in the Residuals vs Fitted plot suggest that homoscedasticity (constant variance of residuals) holds. There is no clear pattern, which is a positive sign, as it indicates that the variance of residuals remains roughly constant across fitted values. 

 **Q-Q Plot (Log-Transformed Residuals)**: In the Q-Q plot, the points being near the diagonal line indicates that the residuals are approximately normally distributed, though there are slight deviations. Some points are slightly above and some slightly below the diagonal, which is not alarming. It suggests that the residuals are close to normal, but there may still be slight skewness or kurtosis in the distribution. The transformation improved the normality of residuals compared to before the transformation, but perfect normality has not yet been achieved. Still, the deviations from the diagonal are quite small, meaning the assumption of normality is reasonably met.

 Now let's check for outliers in the regression model, we can use several approaches:
 
**Standardized Residuals**: If any residuals have an absolute value greater than 2 (or 3), they are considered potential outliers.

**Leverage**: Points with high leverage (above 2 times the mean leverage) can influence the regression model significantly.

**Cook’s Distance**: Cook's distance identifies data points that have an outsized impact on the regression coefficients.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Calculate standardized residuals
standardized_residuals = log_model.get_influence().resid_studentized_internal

# Flagging outliers based on standardized residuals (e.g., > 2 or < -2)
outliers_standardized = np.where(np.abs(standardized_residuals) > 2)[0]

# Leverage values
leverage = log_model.get_influence().hat_matrix_diag

# Flagging points with high leverage (e.g., > 2 * mean leverage)
high_leverage = np.where(leverage > 2 * np.mean(leverage))[0]

# Cook's Distance
cooks_d = log_model.get_influence().cooks_distance[0]

# Flagging points with high Cook's distance (e.g., > 4/n)
high_cooks_d = np.where(cooks_d > 4 / len(data))[0]

# Plot leverage vs. Cook's Distance to visualize the influential points
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(leverage, cooks_d)
ax.set_xlabel('Leverage')
ax.set_ylabel('Cook\'s Distance')
ax.set_title('Leverage vs Cook\'s Distance')

# Marking the influential points
ax.scatter(leverage[high_cooks_d], cooks_d[high_cooks_d], color='red', label='Influential Points')
ax.legend()

plt.show()

# Display outliers and influential points
print(f"Outliers based on standardized residuals: {outliers_standardized}")
print(f"High leverage points: {high_leverage}")
print(f"High Cook's Distance points: {high_cooks_d}")
```

These points are farther away from the mean of the independent variables (GDP and population), meaning they have more potential to influence the regression model. High leverage does not automatically mean these points are outliers, but their influence on the model is stronger, so it’s important to assess them carefully.

## Energy Consumption Simulation Dashboard

To enhance the understanding of the relationship between energy consumption, GDP, and population, I have developed an interactive dashboard. This tool allows users, including public organizations, researchers, and policymakers, to simulate energy consumption based on GDP and population data from various countries.

### Key Features:
- **Interactive Inputs**: Users can input different values for GDP and population to simulate energy consumption scenarios.
- **Visualized Results**: The dashboard visualizes the impact of GDP and population on energy consumption through intuitive graphs and charts.
- **Real-time Feedback**: As users adjust the input parameters, the dashboard dynamically updates the results, providing instant feedback on how energy consumption changes.

### Purpose:
The primary goal of this dashboard is to provide an accessible and user-friendly tool for understanding the energy dynamics between economic factors and population size. This can be especially useful for decision-makers in energy policy, urban planning, and sustainability efforts.

This dashboard helps facilitate a deeper understanding of how these variables interact and provides insight into energy consumption trends across different levels of economic development and population sizes.


```python
!pip install dash plotly
import dash
from dash import dcc, html
import plotly.graph_objects as go
import pandas as pd
from dash.dependencies import Input, Output

# Sample data to train the energy consumption model (can be expanded or modified)
data = {
    'energy': [30633, 58894, 1211 , 27049 , 26405 , 233907, 20119 , 1124 , 11053 , 192453, 456, 84416, 26086 , 12080, 88148, 45132, 23234, 32619, 268056, 352677],
    'gdp': [45451, 62049, 2003, 34540, 28388, 358675, 38039, 1331, 11788, 265863, 1251, 82804, 27914, 30642, 159602, 59350, 42238, 91946, 279191, 428888],
    'population': [11832049, 6445481, 10900555, 5961249, 83445000, 1374687, 5343805, 10397193, 48610458, 68401997, 3861967, 58989749, 933505, 1871882, 2885891, 672050, 9584627, 563443, 17942942, 9158750]
}
df = pd.DataFrame(data)

# Calculate the relationship between energy consumption, GDP, and population
# Here we assume a simple linear regression, but you could use more complex models.
# We can fit a simple linear model energy ~ gdp + population.

# Calculate the energy per capita as a basic approximation for energy consumption.
df['energy_per_capita'] = df['energy'] / df['population']

# Create a simple model based on GDP and population, assuming linear relationships
# For simplicity, let's use the average energy per capita as a multiplier

avg_energy_per_capita = df['energy_per_capita'].mean()

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Energy Consumption Simulator"),
    html.Div([
        html.Label("Enter GDP:"),
        dcc.Input(id='gdp-input', type='number', value=50000),
    ]),
    html.Div([
        html.Label("Enter Population:"),
        dcc.Input(id='population-input', type='number', value=10000000),
]),
    html.Br(),
    html.Div([
        html.Label("Simulated Energy Consumption:"),
        html.Div(id='output-energy-consumption', style={'fontSize': 24}),
    ]),
    html.Br(),
    dcc.Graph(id='gdp-vs-energy-graph'),
])

# Callback function to update energy consumption based on GDP and Population
@app.callback(
    [Output('output-energy-consumption', 'children'),
     Output('gdp-vs-energy-graph', 'figure')],
    [Input('gdp-input', 'value'),
     Input('population-input', 'value')]
)
def update_output(gdp, population):
    # Simulate energy consumption using the average energy per capita
    simulated_energy = avg_energy_per_capita * population

    # Create a graph to show the relationship between GDP and energy consumption
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df['gdp'], y=df['energy'], mode='markers', name='Data'))
    fig.add_trace(go.Scatter(x=[gdp], y=[simulated_energy], mode='markers', name='Simulated Point', marker=dict(color='red', size=12)))

    fig.update_layout(title="Energy Consumption vs GDP",
                      xaxis_title="GDP",
                      yaxis_title="Energy Consumption",
                      showlegend=True)

    # Return the simulated energy value and the updated graph
    return f" {simulated_energy:,.2f} units", fig

# Run the app
if __name__ == "__main__":
    app.run_server(debug=True)
```
